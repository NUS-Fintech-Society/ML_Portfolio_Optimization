# Loading Dataset

```python
#new data scrap
import yfinance as yf
tickers = ['AXP', 'AMGN', 'AAPL', 'BA', 'CAT', 'CSCO', 'CVX', 'GS', 'HD',
       'HON', 'IBM', 'INTC', 'JNJ', 'KO', 'JPM', 'MCD', 'MMM', 'MRK',
       'MSFT', 'NKE', 'PG', 'TRV', 'UNH', 'VZ', 'WBA', 'WMT', 'DIS']
big_df=pd.DataFrame()
for i in tickers:
  df = yf.download(i, start="2001-01-01", end="2020-12-31")
  df['log returns'] = np.log(df['Adj Close']) - np.log(df['Adj Close'].shift(1))
  df['ticker'] = i
  df.dropna(inplace=True)
  big_df=pd.concat([big_df, df])
  time.sleep(1)
```

# Model

```python
#Model
def model(X_train, y_train,X_test,y_test):

  DecisionTree = DecisionTreeRegressor(criterion = 'mse')

  model = DecisionTree.fit(X_train, y_train)

  y_pred = model.predict(X_test)
  mse = metrics.mean_squared_error(y_test, y_pred)
  rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

  return mse, rmse
```

# Results

```python
#Results 
columns = ['ticker', 'MSE', 'RMSE']
df_results = pd.DataFrame(columns=columns)

for ticker in tickers:
  df = big_df[big_df['ticker'] == ticker]
  #splitting 
  df_train, df_test = train_test_split(df, train_size=0.7, test_size=0.3, shuffle=False)
  #training data
  x_train= df_train[['Adj Close','Volume']]
  y_train = df_train[['log returns']]

  #test data
  x_test= df_test[['Adj Close','Volume']]
  y_test = df_test[['log returns']]

  #Results
  ticker_mse, ticker_rmse = model(x_train, y_train , x_test, y_test)
  new = [ticker, ticker_mse, ticker_rmse]
  df_results.loc[df_results.shape[0]] = new
```
